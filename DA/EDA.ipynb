### 2. 매물확인방식 별 허위확률

```python
# 매물확인방식 별 개수
typecheck_counts = df_train["매물확인방식"].value_counts().reset_index()
typecheck_counts.columns = ["매물확인방식", "개수"]
# 매물확인방식 별 허위매물 개수
fraud_typecheck_counts = df_train.query("허위매물여부 == 1")["매물확인방식"].value_counts().reset_index()
fraud_typecheck_counts.columns = ["매물확인방식", "허위매물개수"]
```

```python
merged_typecheck_counts = pd.merge(typecheck_counts, fraud_typecheck_counts, on='매물확인방식', how='left')

# 허위매물 개수가 없는 경우 0으로 채우기
merged_typecheck_counts['허위매물개수'] = merged_typecheck_counts['허위매물개수'].fillna(0)
merged_typecheck_counts["허위확률"] = merged_typecheck_counts["허위매물개수"] / merged_typecheck_counts["개수"]
merged_typecheck_counts
```

```python
# 매물확인방식 별 허위확률

plt.figure(figsize=(8, 6))
sns.barplot(x = "매물확인방식", y = "허위확률", data=merged_typecheck_counts, palette='viridis')
plt.title("매물확인방식 별 허위확률")
plt.xlabel("매물확인방식")
plt.ylabel("허위확률")
plt.show()
```

### 3. 중개사무소 별 허위확률

```python
# 중개사무소 별 개수
agency_counts = df_train["중개사무소"].value_counts().reset_index()
agency_counts.columns = ["중개사무소", "개수"]
# 중개사무소 별 허위매물 개수
fraud_counts = df_train.query("허위매물여부 == 1")["중개사무소"].value_counts().reset_index()
fraud_counts.columns = ["중개사무소", "허위매물개수"]
```

```python
merged_counts = pd.merge(agency_counts, fraud_counts, on='중개사무소', how='left')

# 허위매물 개수가 없는 경우 0으로 채우기
merged_counts['허위매물개수'] = merged_counts['허위매물개수'].fillna(0)
merged_counts["허위확률"] = merged_counts["허위매물개수"] / merged_counts["개수"]
merged_counts
```

```python
merged_counts.sort_values(by="허위확률", ascending=False).head(10)
```

```python
# 중개사무소가 많은 순서대로 top 20개 선정, 허위확률 보기

plt.figure(figsize=(8, 6))
sns.barplot(x="허위확률", y="중개사무소", data=merged_counts.sort_values(by="개수", ascending=False)[:20], palette='viridis')
plt.title("상위 20 중개사무소 별 허위확률")
plt.xlabel("허위확률")
plt.ylabel("중개사무소")
plt.show()
```

### 4. 제공플랫폼 별 정상 vs 허위 비교

```python
df_train["제공플랫폼"].value_counts().reset_index()
```

```python
# 제공플랫폼 별 개수
platform_counts = df_train["제공플랫폼"].value_counts().reset_index()
platform_counts.columns = ["제공플랫폼", "개수"]
# 제공플랫폼 별 허위매물 개수
fraud_platform_counts = df_train.query("허위매물여부 == 1")["제공플랫폼"].value_counts().reset_index()
fraud_platform_counts.columns = ["제공플랫폼", "허위매물개수"]

merged_platform_counts = pd.merge(platform_counts, fraud_platform_counts, on='제공플랫폼', how='left')

# 허위매물 개수가 없는 경우 0으로 채우기
merged_platform_counts['허위매물개수'] = merged_platform_counts['허위매물개수'].fillna(0)
merged_platform_counts["허위확률"] = merged_platform_counts["허위매물개수"] / merged_platform_counts["개수"]
merged_platform_counts
```

```python
# 제공플랫폼 별 허위확률

plt.figure(figsize=(8, 6))
sns.barplot(x = "허위확률", y = "제공플랫폼", data=merged_platform_counts, palette='viridis')
plt.title("제공플랫폼 별 허위확률")
plt.xlabel("허위확률")
plt.ylabel("제공플랫폼")
plt.show()
```

1. 정상매물 87.8%, 허위매물 12.2%
2. 매물확인방식에 따른 보증금 : 전화 방식 이외에 큰 차이 없음.
    + 대체적으로 매물확인방식에서 허위매물이 보증금, 관리비, 월세가 큰 폭을 보임.(높은 값도 있지만 낮은값도 많기에 유의미 X)
#### 허위확률 = 각 데이터별 허위매물 갯수 / 각 데이터별 갯수로 판단
3. 매물확인방식 별 허위확률 : 현장확인(0.131378), 전화확인(0.094241), 서류확인(0.104110)
    + 현장확인 > 서류확인 > 전화확인 순서
    + 값이 다 비슷하기에 매물확인방식을 허위확률지표로 판단하기엔 어려움.
4. 중개사무소 별 허위확률
    + 중개사무소 개수가 많아서 가장 많이 거래한 top20개 중개사무소를 찾아봄.
    + 가장 많은 중개를 한 중개사무소(G52Iz8V2B9) 허위매물 0개

5. 제공플랫폼 별 허위확률
    + 가장 높은 허위확률 : D 플랫폼
    + 가장 낮은 허위확률(M 플랫폼 제외) : F 플랫폼
